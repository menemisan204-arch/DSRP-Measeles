Fix Column

```python
cases_month["year"] = pd.to_numeric(cases_month["year"], errors="coerce")
cases_year["year"] = pd.to_numeric(cases_year["year"], errors="coerce")
```

V1: Total Measle Cases per Year

```python
import matplotlib.pyplot as plt
#Sum measles cases for each year
measles_cases_year = cases_year.groupby("year")["measles_total"].sum()

plt.figure(figsize=(12, 6))
plt.plot(measles_cases_year.index, measles_cases_year.values, marker="o")

plt.title("Global Measles Cases by Year")
plt.xlabel("Year")
plt.ylabel("Total Measles Cases")
plt.grid(True)
plt.show()


```

V2: Total Rubella Cases by Year

```python
rubella_by_year = cases_year.groupby("year")["rubella_total"].sum()
plt.figure(figsize=(12, 6))
plt.plot(rubella_by_year.index, rubella_by_year.values, marker="o")
plt.title("Global Rubella Cases by Year")
plt.xlabel("Year")
plt.ylabel("Total Rubella Cases")
plt.grid(True)
plt.show()
```

V3: Comparison

```python
measles_by_year = cases_year.groupby("year")["measles_total"].sum()
rubella_by_year = cases_year.groupby("year")["rubella_total"].sum()

plt.figure(figsize=(12, 6))

plt.plot(measles_by_year.index, measles_by_year.values, marker="o", label="Measles")
plt.plot(rubella_by_year.index, rubella_by_year.values, marker="o", label="Rubella")

plt.title("Comparison of Measles and Rubella Cases by Year")
plt.xlabel("Year")
plt.ylabel("Total Cases")
plt.legend()
plt.grid(True)
plt.show()
```

Top 10 countries with the most measle cases.

```python
top_measles = (cases_year.groupby("country")["measles_total"].sum().sort_values(ascending=False).head(10))
print(top_measles)
```

Vizualize

```python
plt.figure(figsize=(10, 6))
top_measles.plot(kind="bar", color="skyblue")

plt.title("Top 10 Countries with the Most Measles Cases")
plt.xlabel("Country")
plt.ylabel("Total Measles Cases")
plt.xticks(rotation=45)

plt.show()
```

Top 10 countries with the most rubella cases.




```python
top_rubella = (cases_year.groupby("country")["rubella_total"].sum().sort_values(ascending=False).head(10))
print(top_rubella)
```

```python
plt.figure(figsize=(10, 6))
top_rubella.plot(kind="bar", color="skyblue")

plt.title("Top 10 Countries with the Most Rubella Cases")
plt.xlabel("Country")
plt.ylabel("Total Rubella Cases")
plt.xticks(rotation=45)

plt.show()
```

Compare Cases by Region

```python
region_cases = cases_year.groupby("region")[["measles_total", "rubella_total"]].sum()
print(region_cases)

region_cases.plot(kind="bar", stacked=True, figsize=(10, 6))

plt.title("Cases by Region")
plt.xlabel("Region")
plt.ylabel("Total Cases")

plt.show()
```
